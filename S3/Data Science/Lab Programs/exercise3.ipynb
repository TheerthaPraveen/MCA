Draw a line in a diagram from position (1,3) to (2,10) the to (6,12) and finally to position (18,20)

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,6,18]
y=[3,10,12,20]
plt.figure(figsize=(8,6))
plt.plot(x,y,linestyle=':',color='red',linewidth=2)
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Line and points plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

Program to draw a line using given axis values taken from a text file,with suitable label in the x axis,y axis and a title

```python
import matplotlib.pyplot as plt
def read_data(filename):
    x=[]
    y=[]
    with open(filename,'r')as file:
        for line in file:
            if line.strip():
                parts=line.split()
                if len(parts)==2:
                    xi,yi=map(float,parts)
                    x.append(xi)
                    y.append(yi)
    return x,y
x_values,y_values=read_data('ex3.3.txt')
plt.plot(x_values,y_values,marker='o',linestyle='-',color='blue')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Line plot from file data')
plt.grid(True)
plt.show()

```

Program to plot two or more lines on same plot with suitable legends of each line

```python
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[2,3,5,7,11]
x2=[1,2,3,4,5]
y2=[1,4,6,8,10]
plt.plot(x1,y1,marker='o',linestyle='-',color='blue',label='Line 1')
plt.plot(x2,y2,marker='s',linestyle='--',color='red',label='Line 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.grid(True)
plt.show()
```

Program to create multiple plots

```python
x=np.linspace(0,10,100)
fig,axs=plt.subplots(2,2,figsize=(5,4))
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('sinewave')
axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('cosinewave')
axs[1,0].plot(x,np.tan(x))
axs[1,0].set_title('tangentwave')
axs[1,1].plot(x,np.exp(x))
axs[1,1].set_title('exponentialwave')
plt.tight_layout()
plt.show()


```

Program to create bar plot of scores by groups and gender.

```python
groups=['A','B','C','D','E']
men_means=[22,30,35,35,26]
woman_means=[25,32,30,35,29]
x=np.arange(len(groups))
width=0.35
fig,ax=plt.subplots(figsize=(8,5))
bars1=ax.bar(x-width/2,men_means,width,label='Men',color='blue')
bars2=ax.bar(x+width/2,woman_means,width,label='Woman',color='pink')
ax.set_xlabel('group')
ax.set_ylabel('scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()
plt.show()
```

Program to create a pie chart (popularity of Programming,languages )

Program to create a pie chart of gold medal acheivements of five most successful countries in 2016 Summer olympics

```python
import pandas as pd
data=pd.read_csv('ex3.9.csv')
top=data.sort_values(by='gold_medal',ascending=False).head(5)
plt.figure(figsize=(4,4))
plt.pie(top['gold_medal'],labels=top['country'],autopct='%1.1f%%',startangle=90,shadow=True)
plt.title('Gold medal acheivements of top 5 countries 2016 december olympics')
plt.show()
```

Program to draw to scatter plot

```python
import matplotlib.pyplot as plt

math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]

indices = list(range(len(math_marks)))

plt.figure(figsize=(6,4))
plt.scatter(indices, math_marks, color='red', marker='o', label='Math Marks')
plt.scatter(indices, science_marks, color='green', marker='o', label='Science Marks')

plt.title('Math and Science Marks Comparison')
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.legend()
plt.show()
```
