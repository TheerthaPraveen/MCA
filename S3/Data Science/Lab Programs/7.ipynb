Decision tree 

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
iris=pd.read_csv("ex7.csv")
print(iris)
x=iris[['sore_throat','fever','swollen_glands','congestion','headache']]
y=iris[['diagnosis']]
fname=['sore_throat','fever','swollen_glands','congestion','headache']
tname=['Allergy','Cold','Strep_throat']
clf=DecisionTreeClassifier(criterion="entropy")
clf=clf.fit(x,y)
newcase=[[0,0,1,0,0]]
y_new=clf.predict(newcase)
print(y_new)
plt.figure()
tree.plot_tree(clf,fontsize=10,filled=True,rounded=True,class_names=tname,feature_names=fname)
plt.show()
```
